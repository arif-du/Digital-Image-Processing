# **Loading, displaying, saving image**

Import libraries

```python
import matplotlib.pyplot as plt
import imageio.v3 as iio
```

Load the image

```python
image=iio.imread('brain.jpg')
```

Display the image

```python
plt.imshow(image)
plt.axis('off')
plt.title("Brain MRI")
plt.show()
```

Save the image

```python
output_path='saved_brain.png'
plt.imsave(output_path,image)
print(f"Image saved successfully at {output_path}")
```

# **Conversion from one image mode to another**

import libraries

```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# Load the image
image=Image.open('brain.jpg')
plt.imshow(image)
plt.axis('off')
plt.show()
```

Convert to grayscale

```python
grayscale_image=image.convert('L')
plt.imshow(grayscale_image)
```

Convert to RGBA

```python
rgba_image=image.convert('RGBA')
plt.imshow(rgba_image)
```

Convert to CMYK

```python
cmyk_image=image.convert('CMYK')
plt.imshow(cmyk_image)
```

# **Image Manipulation**

image size, mode and format determination

```python
# Load the image
image=Image.open('brain.jpg')
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
print(f"Width: {image.width}, Height: {image.height}, Mode: {image.mode}, Format: {image.format}")
```

Cropping image

```python
crop_box=(50,50,100,100)
cropped_image=image.crop(crop_box)
plt.imshow(cropped_image)
```

Resizing image

```python
new_size=(30,30)
resized_image=image.resize(new_size)
plt.imshow(resized_image)
```
